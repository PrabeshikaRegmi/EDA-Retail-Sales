```python
df1['Year']=df1['Date'].dt.year
df1['Month']= df1['Date'].dt.month
```

```python
monthly_sales= df1.groupby(['Year','Month']).agg({'Total Amount': 'sum'}).reset_index()
print(monthly_sales)
```

```python
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))
monthly_sales.set_index('Date', inplace=True)
print(monthly_sales)
```

Visualize the sales trends over time

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index,monthly_sales['Total Amount'], marker= 's')
plt.title('Monthly sales over time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

Customer and Product Analysis

```python
gender_analysis= df1.groupby('Gender').agg({'Total Amount':['sum','mean'],'Quantity':['sum','mean']}).reset_index()
age_analysis= df1.groupby('Age').agg({'Total Amount':['sum','mean'],'Quantity':['sum','mean']}).reset_index()

print(gender_analysis)
print(age_analysis)
```

```python
product_gender_analysis= df1.groupby(['Product Category','Gender']).agg({'Total Amount':'sum','Quantity':'sum'}).reset_index()
product_age_analysis= df1.groupby(['Product Category','Age']).agg({'Total Amount':'sum','Quantity':'sum'}).reset_index()
print(product_age_analysis)
print(product_gender_analysis)
```

Plotting total sales by gender

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Gender',y=('Total Amount','sum'), data=gender_analysis, hue='Gender',palette=['blue','red'], legend= True)
plt.title('Total sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.ylim(210000,240000)
plt.show()
```
